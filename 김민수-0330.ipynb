```python
import pandas as pd

data = {'이름' : ['Alice','Bob','Charlie'], '나이' : [25,30,22], '성별' : ['여','남','남']}
df = pd.DataFrame(data, index=['A','B','C'])
print(df)
```

```python

# 첫번째 행 선택 
first_row = df.iloc[0]
print(first_row)
# 두번째 열 (나이) 선택
age_column = df.iloc[:, 1]
print(age_column)
# 첫 번째 행부터 두 번째 행까지, 모든 열 선택
subset = df.iloc[0:2, :]
print(subset)

```

```python
# '이름' 열 선택
name_column = df.loc[:, '이름']
print(name_column)

# 'A'행 선택
row_A = df.loc['A', :]
print(row_A)

# 나이가 25이상인 행 선택
selected_rows = df.loc[df['나이'] >= 25]
print(selected_rows)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 팁 데이터 불러오기
tips = sns.load_dataset('tips')

# 또는 데이터의 처음 5개 행만 출력
print(tips.head())

# 팁 금액 분포 시각화
sns.histplot(data=tips, x='tip', kde=True) # kde : 커널 밀도 추정 곡선
plt.show()
```

```python
# 성별에 따른 팁 금액 비고 (박스플롯)
sns.boxplot(x='sex', y='tip', data=tips)
plt.show()
```

```python
# 팁 금액과 총 지출액 간의 관계 (산점도)
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# 데이터 불러오기
df = pd.read_csv('dataset/diabetes.csv')

# 데이터 정보 확인
print(df.info())

# 1. Violinplot : Qutcome에 따른 Glucose 분포 시각화
plt.Figure(figsize=(8, 6))
sns.violinplot(x='Outcome', y='Glucose', data=df) # 팔레트 종류 : deep, pastel, bright, dark, muted
plt.title('Violin Plot of Glucose by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose Level')
plt.show()
```

```python
# Palrplot : 주요 변수들 간의 관계 시각화
plt.figure(figsize=(10, 10))
sns.pairplot(df, vars=['Glucose','BMI','Age'], hue='Outcome', palette='coolwarm')
plt.suptitle('Pairplot of Key Variables', y=1.02)
plt.show()
```

```python
# Lmplot : Age 와 BMI 간의 회귀 관계 시각화
sns.lmplot(x='Age', y='BMI', data=df, hue='Outcome', palette='Set2', height=7, aspect=1.5)
plt.title('Linear Regression Plot of Age vs BMI by Outcome')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```
